## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
data_1 = pd.read_csv('ml-latest-small/ratings.csv')
data_1.head()
```

```python
data_2 = pd.read_csv('ml-latest-small/movies.csv')
data_2.head()
```

```python
data = pd.merge(data_1, data_2, on = 'movieId', how = 'left')
data.head()
```

```python
filtered_data = data [data['rating'] == 5.0]
filtered_data.head()
```

```python
filtered_data['movieId'].value_counts().head()
```

```python
filtered_data[filtered_data['movieId'] == 318].head()
```

### Ответ: 
фильм с наибольшим количеством оценок 5.0 (153 оценки) - Shawshank Redemption, The (1994), с movieId = 318

## Задача 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
df = pd.read_csv('power.csv')
df.head()
```

```python
baltic_df = df[(df['country'] == 'Lithuania') | (df['country'] == 'Latvia') | (df['country'] == 'Estonia')]
baltic_df.head()
```

```python
final_data = baltic_df[((baltic_df['category'] == 4) | (baltic_df['category'] == 12) | (baltic_df['category'] == 21)) & (baltic_df['year'] >= 2005) & (baltic_df['year'] <= 2010) & (baltic_df['quantity'] >= 0)]
final_data.head(100)
```

```python
final_data['quantity'].sum()
```
